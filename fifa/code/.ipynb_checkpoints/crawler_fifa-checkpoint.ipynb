```python
from multiprocessing import Pool
import pandas as pd 
import os.path
from pandas.io.json import json_normalize
import urllib.request
import requests
from bs4 import BeautifulSoup
import os.path

class Fifa:
    
    def __init__(self):
        self.url = 'https://sofifa.com/'
        self.columns = ['pid', 'nome', 'idade', 'foto', 'nacionalidade', 'bandeira', 'geral', 'potencial', 'clube', 
                       'logo_clube', 'valor', 'salario', 'especial']
        self.data_player = pd.DataFrame()
        self.path_file = 'full_player_data.csv'
        self.path_file_attr = 'full_player_data_attr.csv'
        
    def get_player_file(self):
        self.get_player()
        self.data_player.drop_duplicates()
        self.data_player.to_csv(self.path_file, encoding='utf-8')

    
    def get_player(self, start_page = 0, end_page = 24000):
        base_url = "https://sofifa.com/players?hl=pt-BR&offset="        
        for offset in range(start_page, end_page, 60):
            url = base_url + str(offset)
            source_code = requests.get(url)
            plain_text = source_code.text
            soup = BeautifulSoup(plain_text)
            table_body = soup.find('tbody')
            for row in table_body.findAll('tr'):
                td = row.findAll('td')
                picture = td[0].find('img').get('data-src')
                pid = td[0].find('img').get('id')
                nationality = td[1].find('a').get('title')
                flag_img = td[1].find('img').get('data-src')
                name = td[1].findAll('a')[1].text
                age = td[2].text
                overall = td[3].text.strip()
                potential = td[4].text.strip()
                club = td[5].find('a').text
                club_logo = td[5].find('img').get('data-src')
                value = td[6].text.strip()
                wage = td[7].text.strip()
                special = td[8].text.strip()
                player_data = pd.DataFrame([[pid, name, age, picture, nationality, flag_img, overall, 
                                          potential, club, club_logo, value, wage, special]], columns=self.columns)
                #player_data = pd.concat([player_data, self.get_player_attr(pid)],axis=1,join='inner',keys='pid')
                self.data_player = self.data_player.append(player_data, ignore_index=True)
        return self.data_player
    
    def get_player_attr_file2(self):
        
        if not os.path.exists(self.path_file):
            print('Arquivo de jogadores não existe. Execute o método "get_player_file" para criar.')
            return
        
        self.data_player = pd.read_csv(self.path_file)        
        self.data_player.drop_duplicates()
        
        data_atr = pd.DataFrame()
        for index, row in self.data_player.iterrows():
            data_atr = data_atr.append(self.get_player_attr(row['pid']))
            data_atr['pid'] = row['pid']
        
        data_atr.to_csv('full_player_data_attr.csv', encoding='utf-8')
        
    def get_player_attr_file(self):
        if not os.path.exists(self.path_file):
            print('Arquivo de jogadores não existe. Execute o método "get_player_file" para criar.')
            return
        
        data_attr = pd.DataFrame()
        data_player = pd.read_csv(self.path_file)
        if os.path.exists(self.path_file_attr):
            data_attr = pd.read_csv(self.path_file_attr, encoding='utf-8')['pid']
            data_attr = data_player[not data_player.isin(data_attr['pid'])]['pid']
        else:
            data_attr = data_player['pid']
            data_attr.drop_duplicates()

        data_temp = pd.DataFrame()
        for row in data_attr:
            data_temp = data_temp.append(self.get_player_attr(row))
            data_temp['pid'] = row
            data_temp.to_csv(self.path_file_attr, encoding='utf-8')
            
    def get_player_attr(self, pid):
        player_data_url = 'https://sofifa.com/player/%s?hl=pt-BR'
        dict_prop ={}
        url = player_data_url % pid
        soup = BeautifulSoup(requests.get(url).text)            
        player = soup.find_all('ul',class_='bp3-text-overflow-ellipsis pl')
        categories = soup.select('ul[class=pl]')
        dict_prop['pid'] = pid
        if len(player) > 0:
            dict_prop['perna'] = player[0].find_all('li')[0].text.replace('Perna boa','')
        else:
            dict_prop['perna'] = ''
        player = soup.find_all('ul',class_='bp3-text-overflow-ellipsis pl text-right')
        if len(player) > 0:
            dict_prop['posicao'] = player[0].find_all('li')[2].text.replace('Posição','')
        else:
            dict_prop['posicao'] = ''
        dict_prop['data_fifa'] = soup.find_all('h1')[0].text
        for category in categories:
            skills = category.findAll('li')
            for skill in skills:            
                prop = skill.text.strip().split(' ')
                if len(prop) > 1:
                    dict_prop[prop[1].lower()] = prop[0]

        return json_normalize(dict_prop)
```
