```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
datasus = pd.read_csv(r'data\datasus.zip')
```

```python
df_top10 = datasus.groupby('NOMMUN')['NOMMUN'].count().reset_index(name='OBITOS').sort_values(by='OBITOS',
                                                                                             ascending=False).head(10)
```

```python
plt.figure(figsize=(11,7))
plt.title('Top 10 municípios com óbitos')
b = sns.barplot(x='NOMMUN',y='OBITOS',data=df_top10)
b.set_xticklabels(b.get_xticklabels(), rotation=45);
b.set_ylabel("óbitos")
b.set_xlabel("municípios");
```

```python
df_sexo_estado =  datasus.groupby(['NMESTCIV','NMSEXO'])['NMSEXO'].count().reset_index(name='TOTAL')
```

```python
sns.set(style="whitegrid")
b = sns.catplot(data=df_sexo_estado,x='NMESTCIV',y='TOTAL',hue='NMSEXO',kind='bar')
b.set_ylabels("óbitos")
b.set_xlabels("");
for ax in b.axes.ravel():
    ax.set_xticklabels(ax.get_xticklabels(), rotation=90);
```

```python
df_obito_masculino = datasus[datasus.NMSEXO == 'MASCULINO'].groupby('NMCID')['NMCID'].count().reset_index(name='TOTAL')
```

```python
df_obito_masculino = df_obito_masculino.sort_values(by='TOTAL',ascending=False).head(10)
```

```python
plt.figure(figsize=(11,7))
b = sns.barplot(data=df_obito_masculino, x='TOTAL',y='NMCID',orient='h')
plt.title('Top 10 mortes do sexo masculino')
b.set_xticklabels(b.get_xticklabels(), rotation=90);
b.set_ylabel("óbitos")
b.set_xlabel("Total");
```

```python
df_obito_feminino = datasus[datasus.NMSEXO == 'FEMININO'].groupby('NMCID')['NMCID'].count().reset_index(name='TOTAL')
```

```python
df_obito_feminino = df_obito_feminino.sort_values(by='TOTAL',ascending=False).head(10)
```

```python
plt.figure(figsize=(11,7))
b = sns.barplot(data=df_obito_feminino, x='TOTAL',y='NMCID',orient='h')
plt.title('Top 10 mortes do sexo feminino')
b.set_xticklabels(b.get_xticklabels(), rotation=90);
b.set_ylabel("óbitos")
b.set_xlabel("Total");
```

```python
df_x95_masculino = datasus[(datasus.CODCID.str.contains('X95')) & (datasus.NMSEXO == 'MASCULINO')]
```
